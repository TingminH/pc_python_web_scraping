```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager

# 建立 Service 物件，指定 chromedriver.exe 的路徑
service = Service('./chromedriver.exe')

# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.104.com.tw/")
```

```python
#本來要在console自己寫JS 但這邊可以直接透過Selenium 直接寫然後執行
# driver.execute_script("alert('Test Selenium')")

#捲動到網頁底部
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
# 寫法 1 - 使用 CLASS_NAME
# 先搜尋到你要的元素(by class name) 然後自動輸入你要的值(send keys)
driver.find_element(By.CLASS_NAME, "form-control").send_keys("Python")
```

```python
#把剛才自動填入的字清除掉(找到清除鍵的位置)
#因為用CSS 要用. !
driver.find_element(By.CSS_SELECTOR, "i.jb_icon_delete.input-group__input-clear").click()
```

```python
# 寫法 2 - 使用 CSS_SELECTOR 找父層後再找子層
driver.find_element(By.CSS_SELECTOR, "div.input-group.input-group--search > input.form-control").send_keys("Python")
```

```python
# 寫法 3 - 使用 CSS_SELECTOR 使用屬性精準匹配
driver.find_element(By.CSS_SELECTOR, "input[placeholder='關鍵字(例：工作職稱、公司名、技能專長...)']").send_keys("Python")
```

```python
# 點擊搜尋按鈕 #CSS selector不能有空格
driver.find_element(By.CSS_SELECTOR, "button.nav-search-btn").click()
```

### 查找不同地區的資料

```python
#老師的
# 取得所有區域選項的 value 以提供後續訪問各地區使用
"""
<a class="category-picker-checkbox">
    <span class="category-picker-checkbox-input">
        <input type="checkbox" class="checkbox-input" value="6001001000">
    </span>
    <span class="children">台北市</span>
</a>
"""

area_elements = driver.find_elements(By.CSS_SELECTOR, ".category-picker-checkbox")
```

```python
# ---- 測試第一筆 ----
print(area_elements[0].find_element(By.CSS_SELECTOR, "input").get_attribute("value"))
print(area_elements[0].find_element(By.CSS_SELECTOR, ".children").text)
```

```python
# ---- 取得所有區域的 dict ----
district_value = {}

for area_element in area_elements:
    area_value = area_element.find_element(By.CSS_SELECTOR, "input").get_attribute("value")
    area_name = area_element.find_element(By.CSS_SELECTOR, ".children").text
    district_value[area_name] = area_value
```

```python
# 檢視區域的 mapping
district_value
```
