```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager

# 建立 Service 物件，指定 chromedriver.exe 的路徑
service = Service('./chromedriver.exe')

# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(options=options)
driver.get("https://www.1111.com.tw/")

# 使用 CSS_SELECTOR 找父層後再找子層 輸入關鍵字
driver.find_element(By.CSS_SELECTOR, "div.relative.w-full > input.keyword-input").send_keys("Python")

# 點擊搜尋按鈕 
driver.find_element(By.CSS_SELECTOR, "span.text-white").click()

# 取得所有區域選項的 value 以提供後續訪問各地區使用
"""
<div class="flex flex-col h-fit w-full"><span class="text-main font-medium px-4 py-2 leading-[24px]">北部</span>
    <div>
        <ul class="flex-col-wrap">
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between hover:bg-gray-100">
                台北市 <!----></li>
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between bg-blue-000 hover:bg-blue-000">
                新北市 <!----></li>
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between hover:bg-gray-100">
                基隆市 <!----></li>
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between hover:bg-gray-100">
                桃園市 <!----></li>
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between hover:bg-gray-100">
                新竹市 <!----></li>
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between hover:bg-gray-100">
                新竹縣 <!----></li>
            <li
                class="flex flex-row items-center min-h-10 w-full cursor-pointer py-2 px-4 justify-between hover:bg-gray-100">
                苗栗縣 <!----></li>
        </ul>
    </div>
</div>

"""

area_elements = driver.find_elements(By.CSS_SELECTOR, "div.flex.flex-row.items-center")

```
