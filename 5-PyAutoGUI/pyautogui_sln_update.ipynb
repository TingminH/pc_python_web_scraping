### 小畫家繪製漸層螺旋圖案


```python
import pyautogui
import time

"""
取得起始點的方式
1. time.sleep(2) 搭配 pyautogui.position() 取得滑鼠位置
2. 在 cmd 使用 python -m pyautogui 進入互動模式
"""

# 設定起始點
start_point = (500, 350)
# 移動滑鼠到起始點
pyautogui.moveTo(start_point[0], start_point[1])

# 設定大小
distance = 200
# 按下滑鼠左鍵
pyautogui.mouseDown()

while distance > 0:
    # 向右拖曳
    pyautogui.dragRel(distance, 0, duration=0.1)
    distance -= 10
    # 向下拖曳
    pyautogui.dragRel(0, distance, duration=0.1)
    # 向左拖曳
    pyautogui.dragRel(-distance, 0, duration=0.1)
    distance -= 10
    # 向上拖曳
    pyautogui.dragRel(0, -distance, duration=0.1)

# 放開滑鼠左鍵
pyautogui.mouseUp()
```

### 螢幕截圖 與 圖片辨識

```python
import pyautogui
"""
小算盤
擷取數字圖片後自動點擊
"""

# 要擷取數字的大小
num_size = 40
# 數字之間的水平與垂直距離
x_distance = 80
y_distance = 50

# 起始位置 (從數字1的位置開始)
start_location = (20, 410)

for i in range(3):
    # 擷取數字 1~3        
    region = (
        start_location[0] + i * x_distance, 
        start_location[1], 
        num_size, 
        num_size
    )
    pyautogui.screenshot(f"num_{i+1}.png", region=region)

    # 擷取數字 4~6
    region = (
        start_location[0] + i * x_distance, 
        start_location[1] - y_distance, 
        num_size, 
        num_size
    )
    pyautogui.screenshot(f"num_{i+4}.png", region=region)

    # 擷取數字 7~9
    region = (
        start_location[0] + i * x_distance, 
        start_location[1] - y_distance*2, 
        num_size, 
        num_size
    )
    pyautogui.screenshot(f"num_{i+7}.png", region=region)

region = (
    start_location[0] + x_distance, 
    start_location[1] + y_distance, 
    num_size, 
    num_size
)
pyautogui.screenshot('num_0.png', region=region)

print("圖片擷取完成")
```

```python
def click(num):
    location = pyautogui.locateCenterOnScreen(f'num_{num}.png', confidence=0.9)
    if location:
        # center = pyautogui.center(location)
        pyautogui.click(location)

click(5)
click(2)
click(0)
```

```python
import pyautogui
import keyboard
# pyautogui.PAUSE = 0.1  # 每個 PyAutoGUI 函式呼叫後暫停 1 秒
'''
打擊遊戲 - 魔術鋼琴鏄
https://html5.gamedistribution.com/rvvASMiM/44fbc5e5c6e54ac9985d1c81f1fb8121/index.html
'''
point_list = [
    (1000, 900),
    (1200, 900),
    (1400, 900),
    (1600, 900),
]

def click():
    for point in point_list:
        color = pyautogui.pixel(point[0], point[1])

        if color[0] == 0 and color[1] == 0:
            pyautogui.mouseDown(point[0], point[1])
            pyautogui.mouseUp()

while True:
    if keyboard.is_pressed('Ctrl'):
        click()
    elif keyboard.is_pressed('q'):
        break
```

```python
import pyautogui
import keyboard
"""
太鼓達人
https://cjdgrevival.com/
blue_rgb = (101, 189, 187)
orange_rgb = (243, 71, 40)
orange_rgb_alt = (248, 119, 0)
yellow_rgb = (243, 181, 0)
"""

# 顏色範圍設定
blue_range = {
    "r": (95, 105),
    "g": (180, 195),
    "b": (180, 190)
}

orange_range = {
    "r": (240, 250),
    "g": (70, 75),
    "b": (35, 45)
}

orange_alt_range = {
    "r": (245, 255),
    "g": (110, 125),
    "b": (0, 10)
}

yellow_range = {
    "r": (235, 245),
    "g": (175, 185),
    "b": (0, 10)
}

def in_range(color, range_dict):
    r, g, b = color
    return (range_dict["r"][0] <= r <= range_dict["r"][1] and
            range_dict["g"][0] <= g <= range_dict["g"][1] and
            range_dict["b"][0] <= b <= range_dict["b"][1])

# 偵測的座標
location = (720, 590)

def click():
    pixel = pyautogui.pixel(*location)

    if in_range(pixel, blue_range):
        pyautogui.press('D')

    elif in_range(pixel, orange_range):
        pyautogui.press('F')

    elif in_range(pixel, orange_alt_range):
        for _ in range(3):
            pyautogui.press('D')
            pyautogui.press('F')

    elif in_range(pixel, yellow_range):
        for _ in range(3):
            pyautogui.press('D')
            pyautogui.press('F')

while True:
    if keyboard.is_pressed('Ctrl'):
        click() 
    elif keyboard.is_pressed('Q'):
        print("退出程式")
        break
```
