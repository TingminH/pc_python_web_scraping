### 小畫家繪製漸層螺旋圖案


```python
import pyautogui
import time

"""
取得起始點的方式
1. time.sleep(2) 搭配 pyautogui.position() 取得滑鼠位置
2. 在 cmd 使用 python -m pyautogui 進入互動模式
"""

start_point = (500, 350)
pyautogui.moveTo(start_point)

pyautogui.mouseDown()
distance = 200

while distance > 0:
    pyautogui.dragRel(distance, 0, duration=0.1)
    distance -= 10
    pyautogui.dragRel(0, distance, duration=0.1)
    pyautogui.dragRel(-distance, 0, duration=0.1)
    distance -= 10
    pyautogui.dragRel(0, -distance, duration=0.1)

pyautogui.mouseUp()
```

```python
#Mac 上不行執行上方的code 問GPT改成以下的：

import pyautogui
import time

# 起始點座標（你可以依照自己螢幕調整）
start_point = (500, 350)

# 先把滑鼠移到起始點
pyautogui.moveTo(start_point)

# 按下滑鼠左鍵（開始畫圖）
pyautogui.mouseDown(button='left')

distance = 200  # 一開始邊長 200 像素

while distance > 0:
    # 往右拖曳 distance
    pyautogui.dragRel(distance, 0, duration=0.1, button='left')
    distance -= 10  # 邊長縮短 10

    # 往下拖曳 distance
    pyautogui.dragRel(0, distance, duration=0.1, button='left')

    # 往左拖曳 distance
    pyautogui.dragRel(-distance, 0, duration=0.1, button='left')
    distance -= 10  # 再縮短 10

    # 往上拖曳 distance
    pyautogui.dragRel(0, -distance, duration=0.1, button='left')

# 放開滑鼠左鍵
pyautogui.mouseUp(button='left')
```

```python
import pyautogui
import keyboard
# pyautogui.PAUSE = 0.1  # 每個 PyAutoGUI 函式呼叫後暫停 1 秒
'''
打擊遊戲 - 魔術鋼琴鏄
https://html5.gamedistribution.com/rvvASMiM/44fbc5e5c6e54ac9985d1c81f1fb8121/index.html
'''
point_list = [
    (1000, 900),
    (1200, 900),
    (1400, 900),
    (1600, 900),
]

def click():
    for point in point_list:
        color = pyautogui.pixel(point[0], point[1])

        if color[0] == 0 and color[1] == 0:
            pyautogui.mouseDown(point[0], point[1])
            pyautogui.mouseUp()

while True:
    if keyboard.is_pressed('Ctrl'):
        click()
    elif keyboard.is_pressed('q'):
        break
```
